```python
# íŠ¹ì •ìš”ì†Œê°€ ë¡œë“œë  ë•Œê¹Œì§€ ëŒ€ê¸°(WebDriverWait)
from selenium.webdriver.support.ui import WebDriverWait
# ëŒ€ê¸° ì¡°ê±´
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get("https://www.naver.com")

# nameì˜ ê°’ì´ qì¸ ìš”ì†Œê°€ ë¡œë“œë  ë•Œê¹Œì§€ 10ì´ˆê°„ ê¸°ë‹¤ë¦° í›„ ê·¸ ìš”ì†Œë¥¼ ì„ íƒí•¨
searchBox = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.NAME, "query")) # íŠœí”Œí˜•ìœ¼ë¡œ ì§‘ì–´ë„£ì–´ì•¼ í•¨
)
searchBox.send_keys("selenium ëŒ€ê¸° í…ŒìŠ¤íŠ¸")
searchBox.send_keys(Keys.ENTER)
time.sleep(5)

driver.quit()
```

```python
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()
driver.get("https://www.google.com")

# elementì— ë§ˆìš°ìŠ¤ ì˜¤ë²„ ì´ë²¤íŠ¸(move_to_element() ë©”ì†Œë“œë¥¼ ì‚¬ìš©)ë¥¼ ë§Œë“¬
element = driver.find_element("link text", "Gmail")
actions = ActionChains(driver)  # ActionChains ê°ì²´ë¥¼ ìƒˆë¡œ ë§Œë“¬
actions.move_to_element(element).perform()
time.sleep(10)
driver.quit()
```

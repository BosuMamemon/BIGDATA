```python
from bs4 import BeautifulSoup as bs
import requests
import re   # 정규 표현식
import json
```

```python
baseUrl = "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page="
books = []

for i in range(10):
    url = baseUrl + str(i+1)
    res = requests.get(url)
    soup = bs(res.text, "html.parser")

    bookList = soup.findAll("div", {"class": "ss_book_list"})
    for index, book in enumerate(bookList):
        if index % 2 == 0:
            title = book.find("a", {"class": "bo3"}).text
            price = int(book.find("ul").findAll("li")[3].find("span").text.replace(",", ""))
            author = book.find("ul").findAll("li")[2].find("a")
            print((i+1), title, price, author)
            if author != None:
                author = author.text
            else:
                author = book.find("ul").findAll("li")[1].find("a").text
```

```python
import pandas as pd

df = pd.DataFrame(books, columns=["제목", "가격", "저자"])
df.to_csv("../data/알라딘.csv", encoding="utf-8")
```

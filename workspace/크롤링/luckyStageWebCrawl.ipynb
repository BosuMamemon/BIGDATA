```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas
import pymysql
import urllib.request as req
import re
```

```python
result = []

driver = webdriver.Chrome()
driver.get("https://ticket.melon.com/concert/index.htm?genreType=GENRE_CON")
old_window = driver.current_window_handle
time.sleep(2)

last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")    
    time.sleep(2)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

time.sleep(2)

concert_list = driver.find_elements(By.CSS_SELECTOR, "#perf_poster > li")

for index, concert in enumerate(concert_list, 1):
    if concert.find_element(By.CSS_SELECTOR, "a").get_attribute("href") == "https://ticket.melon.com/concert/index.htm?genreType=GENRE_CON#": break
    link = concert.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
    driver.execute_script("window.open(arguments[0]);", link)
    driver.switch_to.window(driver.window_handles[1])
    time.sleep(4)

    concert_num = index
    title = driver.find_element(By.CSS_SELECTOR, "#conts > div > div.wrap_consert_product > div.wrap_consert_cont > div.box_consert_txt > p.tit").text

    poster_src = driver.find_element(By.CSS_SELECTOR, "#conts > div > div.wrap_consert_product > div.wrap_consert_cont > div.box_consert_thumb.thumb_180x254 > img").get_attribute("src")
    pattern = r"[가-힣a-zA-Z0-9]+"
    patterned = re.findall(pattern, title)
    title_patterned = " ".join(patterned)
    req.urlretrieve(poster_src, f"C:/JHM/BIGDATA/data/luckyStage/poster_{title_patterned}.png")
    concert_posters = f"poster_{title_patterned}.png"
    print(concert_posters, "다운로드됨...")

    date = driver.find_element(By.CSS_SELECTOR, "#periodInfo").text.split(" - ")
    start_date = date[0]
    end_date = date[1]
    performance_time = driver.find_element(By.CSS_SELECTOR, "#conts > div > div.wrap_consert_product > div.wrap_consert_cont > div.box_consert_txt > div.box_consert_info > dl.info_left > dd:nth-child(4)").text
    location = driver.find_element(By.CSS_SELECTOR, "#performanceHallBtn > span.place").text
    age_rate = driver.find_element(By.CSS_SELECTOR, "#conts > div > div.wrap_consert_product > div.wrap_consert_cont > div.box_consert_txt > div.box_consert_info > dl.info_right > dd:nth-child(4)").text
    
    url = ""
    img_list = driver.find_elements(By.CSS_SELECTOR, "#conts > div > div:nth-child(3) > div.wrap_detail_left_cont > div.box_img_content img")
    for index, img in enumerate(img_list):
        img_src = img.get_attribute("src")
        req.urlretrieve(img_src, f"C:/JHM/BIGDATA/data/luckyStage/content_{title_patterned}_{index}.png")
        print(f"content_{title_patterned}_{index}.png", "다운로드됨...")
        url += f"content_{title_patterned}_{index}.png" + "-"
    result.append((concert_num, title, url, start_date, end_date, performance_time, location, age_rate, link, concert_posters))
    
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(4)

driver.quit()

df = pandas.DataFrame(result, columns=["concert_num", "title", "concert_content", "start_date", "end_date", "performance_time", "location", "age_rate", "payment_link", "concert_poster"])
df.to_csv("../data/luckyStage.csv", encoding="utf-8")

```

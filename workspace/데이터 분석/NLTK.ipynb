```python
# 정규표현식 토크나이저 객체 생성 (패턴은 워드가 1자 이상인 것)
regTokenizer = RegexpTokenizer("[\w]+")
regToken = regTokenizer.tokenize(emma_raw)
regToken
```

```python
# 형태소 분석

from nltk.stem import PorterStemmer, LancasterStemmer
st1 = PorterStemmer() 
st2 = LancasterStemmer() 
words = ['fly','flies','flying','flew','flown']

print("Porter Stemmer: ", [st1.stem(w) for w in words]) 
print("Lancaster Stemmer:", [st2.stem(w) for w in words])

list1 = []
for w in words: 
    print(st1.stem(w), st2.stem(w)) 
    list1.append(st1.stem(w))
```

```python
#  표제어 추출

from nltk.stem import WordNetLemmatizer

lm = WordNetLemmatizer()
v1 = [lm.lemmatize(w, 'v') for w in words]  # v는 해당 단어를 동사로 생각하라는 뜻
print(v1)

list2 = []
for w in words:
    list2.append(lm.lemmatize(w, 'v')) 
    print(list2)
```

```python
from nltk.tag import pos_tag 
tagged_list = pos_tag(regTokenizer.tokenize(emma_raw)) #(word_tokenize(sentence)) 
tagged_list
```

```python
nltk.help.upenn_tagset('NNP') #품사 확인
```

```python
# 고유 명사 및 명사 추출

noun_list=[t[0] for t in tagged_list if t[1]=='NNP']
noun_list
```

```python
# 불용어 설정
stop_word = ["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear", "A", "No", "Ah", "Oh", "VOLUME", "Between", "Woodhouse"]
name_list = [t[0] for t in tagged_list if t[1] == "NNP" and t[0] not in stop_word]
name_list
```

```python
fd_name = nltk.FreqDist(name_list)
fd_name
```

```python
leaves_raw = nltk.corpus.gutenberg.raw(fileids="whitman-leaves.txt")
# regTokenizer = RegexpTokenizer("[\w]+")
leaves_token = regTokenizer.tokenize(leaves_raw)
leaves_token
```

```python
leaves_tag = pos_tag(leaves_token)
leaves_tag
```

```python
leaves_noun = [t[0] for t in leaves_tag if t[1] == "NN"]
leaves_noun
```

```python
leaves_noun_fd = nltk.FreqDist(leaves_noun)
leaves_noun_fd
```

```python
stopwords = ['d', 's', 'st', 'o', 're']
leaves_noun = [t[0] for t in leaves_tag if t[1] == 'NN' and t[0] not in stopwords]
leaves_noun_fd = nltk.FreqDist(leaves_noun)
leaves_noun_fd
```

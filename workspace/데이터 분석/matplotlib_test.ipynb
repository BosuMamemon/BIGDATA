```python
df = pd.read_excel("./data/fine_dust.xlsx", index_col=0)
df.head()
```

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data_2023.index, data_2023.values, marker="o", color="blue", label="2023년")
plt.title("2023년 미세먼지 농도")
plt.xlabel("날짜")
plt.ylabel("농도")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for year in range(2020, 2024):
    chart_data = df[year]
    plt.plot(chart_data.index, chart_data.values, marker="o", label=f"{year}년")
plt.title("미세먼지 농도")
plt.xlabel("Area")
plt.ylabel("농도")
plt.legend()
plt.show()
```

```python
data4 = df.loc[["부산", "대구", "인천", "경기"], 2020:2024]
data4
```

```python
plt.figure(figsize=(15, 4))
index = np.arange(4)
for year in range(2020, 2024):
    chart_data = data4[year]
    plt.bar(index, chart_data, label=f"{year}년", width=0.2)
    index = index + 0.2
plt.title("미세먼지 농도")
plt.xticks(index-0.5, ["부산", "대구", "인천", "경기"])
plt.ylim(25, 45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
index = np.arange(4)
data4 = df.loc[["대구", "인천", "부산", "경기"], 2017:2023]
for year in range(2017, 2024):
    chart_data = data4[year]
    plt.barh(index, chart_data, label=year, height=0.125)
    index = index + 0.125
plt.ylabel("Area")
plt.xlabel("micrometer")
plt.yticks(index-0.5, ["대구", "인천", "부산", "경기"])
plt.title("미세먼지 어쩌고")
plt.legend()
plt.show()
```

```python
# df로 그리면 matplotlib보다 누적 차트는 더 편할 때도 있음
data4.plot(kind="bar", figsize=(10,4), stacked=True)
data4.plot(kind="bar", figsize=(10,4), stacked=False)
```

```python
plt.figure(figsize=(15, 5))
bottom = np.zeros(4)
for year in range(2020, 2024):
    chart_data = data4[year]
    plt.bar(index, chart_data, label=f"{year}년", bottom=bottom)
    bottom = bottom + chart_data
plt.title("미세먼지 농도")
plt.xticks(index, ["부산", "대구", "인천", "경기"])
plt.legend()
plt.show()
```

```python
health = pd.read_excel("./data/health_screenings_2020_1000ea.xlsx")
health.head()
```

```python
height_data = health["height"]
height_data
```

```python
weight_data = health["weight"]
weight_data
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(height_data, weight_data)
plt.title("키와 몸무게의 관계")
plt.xlabel("키")
plt.ylabel("몸무게")
plt.grid()
plt.show()
```

```python
hw_df = pd.DataFrame({"height": height_data, "weight": weight_data})
hw_df.corr(method="pearson")
```

```python
systolic_data = health["systolic"]  # 수축기 혈압
diastolic_data = health["diastolic"]    # 이완기 혈압
blood_sugar_data = health["blood_sugar"]    # 공복 혈당

plt.figure(figsize=(10, 6))
plt.scatter(systolic_data, diastolic_data, color="red", edgecolors="black", label="systolic * diastolic")
plt.scatter(systolic_data, blood_sugar_data, color="blue", edgecolors="black", label="systolic * blood_sugar")
plt.scatter(diastolic_data, blood_sugar_data, color="green", edgecolors="black", label="diastolic * blood_sugar")

plt.title("아무도모르게다가온이별에대면했을때 또다시혼자가되는게두려워외면했었네")
plt.legend()
plt.grid()
plt.show()
```

```python
health_corr1 = pd.DataFrame({"systolic": systolic_data, "diastolic": diastolic_data})
health_corr2 = pd.DataFrame({"systolic": systolic_data, "blood_sugar": blood_sugar_data})
health_corr3 = pd.DataFrame({"diastolic": diastolic_data, "blood_sugar": blood_sugar_data})

health_corr1.corr("pearson")
```

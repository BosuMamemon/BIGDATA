```python
from nltk import Text
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
# 워드 클라우드 한글 사용 설정
font_path = "./data/fonts/MALGUN.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
kolaw_nouns = Text(okt.nouns(c))
kolaw_nouns
```

```python
kolaw_nouns.plot(20)
```

```python
# FreqDist 객체로 텍스트 빈도수 구함 
kv=kolaw_nouns.vocab() 
kv.items()
```

```python
kv2 = dict()

for key, val in kv.items():
    if(len(key) > 1 and val > 1):
        kv2[key] = val

kv2
```

```python
from wordcloud import WordCloud
from PIL import Image
import numpy as np

mask_arr = np.array(Image.open('./data/agumon.jpg'))

wc = WordCloud(width=800, height=800, background_color='white', random_state=0,
               font_path=font_path, mask=mask_arr, colormap='Accent_r')    # font_path에 한글 폰트 안넣어놓으면 깨짐
plt.axis('off')
plt.imshow(wc.generate_from_frequencies(kv2))
plt.savefig('./data/wc03.png')
```

```python
from collections import Counter

s1 = 'aaabbccdddde'
list1 = ['abc', 'abc', 'aa', 'aa', 'bb', 'cc', 'bb', 'bb']

counter1 = Counter(s1)
counter2 = Counter(list1)

print(counter1)
print(counter2)
```

```python
nouns1 = okt.nouns(c)
nouns1_count1 = Counter(nouns1)
nouns1_count1
```

```python
nouns1_count2=nouns1_count1.most_common(100) 
nouns1_count2
```

```python
count3 = dict()
#단어의 수가 1보다크고, 빈도수도 1보다 큰것을 선택 
for key, value in nouns1_count1.items(): 
    if(len(key) > 1 and value > 1): 
        count3[key] = value

count3
```

```python
wc = WordCloud(width=800, height=800, background_color='white', colormap='Accent_r',
               font_path=font_path, random_state=100, mask=mask_arr)
plt.imshow(wc.generate_from_frequencies(count3))
plt.axis('off')
plt.savefig('./data/wc4.png')
```

1809890.txt를 사용하여 다음을 수행하라
1. Okt 형태소 분석기를 사용하여 명사를 추출하라.
2. collection.Counter를 사용하여 단어의 빈도수를 구하라.
3. 빈도수가 2 이상이고 글자수가 2 이상인 단어를 추출하라.
4. 위 결과를 워드 클라우드로 시각화하라.

```python
raw_text = kobill.open('1809890.txt').read()
okt_text = okt.nouns(raw_text)  #1
okt_count = Counter(okt_text)   #2

result = dict()
for key, val in okt_count.items():
    if(len(key) >= 2 and val >= val):
        result[key] = val   #3

wc = WordCloud(width=1200, height=1200, background_color='white', font_path=font_path)  #4
plt.imshow(wc.generate_from_frequencies(result))
plt.axis('off')
```

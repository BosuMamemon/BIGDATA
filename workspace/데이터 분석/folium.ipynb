```python
tooltips = [[37.567826456,126.9776579], [37.565826456,126.979657]]
lat = 37.56682645618825
lon = 126.97865791770842
popups=['aa','bb']
latlngs = [[37.567826456,126.9776579], [37.565826456,126.979657]]

m = folium.Map(location=[lat, lon], zoom_start=15)
folium.Marker([lat, lon], popup='서울시청').add_to(m)
for i in range(len(latlngs)):
    folium.Marker([latlngs[i][0], latlngs[i][1]], 
                  popup=popups[i], 
                  tooltip=f'lat:{tooltips[i][0]},long:{tooltips[i][1]}').add_to(m)
m
```

```python
N=100
data=np.array(
    [np.random.uniform(low=35, high=39, size=N), 
      np.random.uniform(low=125, high=129,size=N)]).T
data
```

```python
m = folium.Map(location=[lat, lon], zoom_start=10)
for i in range(N):
    folium.Marker([data[i][0], data[i][1]], 
                  popup=f'lat:{data[i][0]}, long:{data[i][1]}',
                ).add_to(m)
m
```

```python
from folium import plugins
popups=[str(i) for i in range(N)]
m = folium.Map(location=[lat, lon], zoom_start=10)
plugins.MarkerCluster(data,popups=popups).add_to(m)
m
```

```python
df= pd.read_csv('data8/CCTV_20190917.csv')
df.head()
```

```python
df1=df.loc[df['카메라대수']>=1,:]
df1.head()
```

```python
df1['소재지도로명주소']= df1['소재지도로명주소'].apply(str.replace(' ', ''))
popups = df1['소재지도로명주소'].tolist()
popups
```

```python
lat_avg=np.array(df1['위도'].tolist()).mean()
lng_avg=np.array(df1['경도'].tolist()).mean()
m = folium.Map(location=[lat_avg, lng_avg], zoom_start=15)
m
```

```python
data=np.array([df1['위도'].tolist(), df1['경도'].tolist()]).T
for i in range(len(data)):
    folium.Marker([data[i][0], data[i][1]], 
        popup=popups[i],
        icon=folium.Icon(color='red',icon='info-sign')).add_to(m)
m
```

```python
m = folium.Map(location=[lat_avg, lng_avg], zoom_start=12)
plugins.MarkerCluster(data, popups=popups).add_to(m)
m.save('data8/CCTV_map.html')
m
```

### 4.2 행정구역별 인구수 데이터 준비하기

```python
population = pd.read_excel('data8/행정구역_시군구_별__성별_인구수_2.xlsx')
population.head() #작업 확인용 출력
```

```python
local_MC_Population = addr_population_merge[ ['시도_x', '군구_x','count', '총인구수 (명)']]
local_MC_Population.head() #작업 확인용 출력
```

### 3. 블록맵으로 시각화하기

```python
BORDER_LINES = [
    [(3, 2), (5, 2), (5, 3), (9, 3), (9, 1)], # 인천
    [ (2, 5), (3, 5), (3, 4), (8, 4), (8, 7), (7, 7), (7, 9), (4, 9),
    (4, 7), (1, 7)], # 서울
    [(1, 6), (1, 9), (3, 9), (3, 10), (8, 10), (8, 9),
    (9, 9), (9, 8), (10, 8), (10, 5), (9, 5), (9, 3)], # 경기도
    [(9, 12), (9, 10), (8, 10)], # 강원도
    [(10, 5), (11, 5), (11, 4), (12, 4), (12, 5), (13, 5),
    (13, 4), (14, 4), (14, 2)], # 충청남도
    [(11, 5), (12, 5), (12, 6), (15, 6), (15, 7), (13, 7),
    (13, 8), (11, 8), (11, 9), (10, 9), (10, 8)], # 충청북도
    [(14, 4), (15, 4), (15, 6)], # 대전시
    [(14, 7), (14, 9), (13, 9), (13, 11), (13, 13)], # 경상북도
    [(14, 8), (16, 8), (16, 10), (15, 10),
    (15, 11), (14, 11), (14, 12), (13, 12)], # 대구시
    [(15, 11), (16, 11), (16, 13)], # 울산시
    [(17, 1), (17, 3), (18, 3), (18, 6), (15, 6)], # 전라북도
    [ (19, 2), (19, 4), (21, 4), (21, 3), (22, 3), (22, 2), (19, 2)], #광주시
    [(18, 5), (20, 5), (20, 6)], # 전라남도
    [ (16, 9), (18, 9), (18, 8), (19, 8), (19, 9), (20, 9), (20, 10)], #부산시
]
```

```python
def draw_blockMap(blockedMap, targetData, title, color):
    whitelabelmin = ( max(blockedMap[targetData]) - min(blockedMap[targetData])) * 0.25 + min(blockedMap[targetData])
    print(whitelabelmin)
    datalabel = targetData
    #print("kk",datalabel)
    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])
    #print(blockedMap)
    mapdata = blockedMap.pivot(index = 'y', columns = 'x', values = targetData)
    #print(mapdata)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    print(masked_mapdata)

    
    plt.figure(figsize = (8, 13))
    plt.title(title)
    plt.pcolor( masked_mapdata, vmin = vmin, vmax = vmax, cmap = color, edgecolor = '#aaaaaa', linewidth = 0.5)

    #지역 이름 표시
    for idx, row in blockedMap.iterrows():
        #annotation 색상지정, whitelabelmin 값보다 크면 white, 작거나 같으면 black로 지정
        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        #광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시, 세종시는 제외
        if row['광역시도'].endswith('시') and not row['광역시도'].startswith('세종'):
            # 광역시도 이름 2자 + 행정구역 이름 마지막 글자 제거하여 dispname 만듬
            dispname = ' {}\n{}'.format(row['광역시도'][:2], row['행정구역'][:-1])
            # 행정구역 이름이 2자보다 작으면 dispname에 행정구역의 마지막 글자를 붙혀 줌
            if len(row['행정구역']) <= 2: 
                dispname += row['행정구역'][-1]
        else:
            dispname = row['행정구역'][:-1] # 행정구역에 마지막 글자 제거
            
        #서대문구, 서귀포시 같이 이름이 3자 이상이면 작은 글자로 표시
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2
            
            plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight = 'bold',fontsize = fontsize, 
                         ha = 'center', va = 'center', color = annocolor, linespacing = linespacing)
        #시도 경계를 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c = 'black', lw = 4)
        
    plt.gca().invert_yaxis()
    plt.gca().set_aspect(1)
    plt.axis('off')

    cb = plt.colorbar(shrink = 1, aspect = 10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.savefig('.\\data8\\' + 'blockMap_' + targetData + '.png')
    plt.show()
```

```python
def draw_blockMap1(blockedMap, targetData, title, color):
    # 라벨의 글자 색상 결정 기준입니다. 데이터 값이 이 값보다 크면 흰색 글자, 아니면 검은색 글자를 사용을 위한 기준
    whitelabelmin = ( max(blockedMap[targetData]) - min(blockedMap[targetData])) * 0.25 + min(blockedMap[targetData])
    datalabel = targetData
    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])
    # x, y 좌표에 따라 targetData 값을 배치한 2차원 매트릭스를 만듬
    mapdata = blockedMap.pivot(index = 'y', columns = 'x', values = targetData)
    # 결측치(NaN)는 표시하지 않도록 마스킹 처리.
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize = (8, 13))
    plt.title(title)

    # 색상 맵으로 시각화합니다 (pcolor는 2D 컬러 플롯을 그립니다).
    # vmin/vmax는 색상 범위.
    # 각 셀의 테두리를 #aaaaaa 회색으로 설정.
    plt.pcolor( masked_mapdata, vmin = vmin, vmax = vmax, cmap = color, edgecolor = '#aaaaaa', linewidth = 0.5)

    #지역 이름 표시
    for idx, row in blockedMap.iterrows():
        #annotation 색상지정, whitelabelmin 값보다 크면 white, 작거나 같으면 black로 지정
        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        #광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시, 세종시는 제외
        if row['광역시도'].endswith('시') and not row['광역시도'].startswith('세종'):
            # 광역시도 이름 2자 + 행정구역 이름 마지막 글자 제거하여 dispname 만듬
            dispname = ' {}\n{}'.format(row['광역시도'][:2], row['행정구역'][:-1])
            # 행정구역 이름이 2자보다 작으면 dispname에 행정구역의 마지막 글자를 붙혀 줌
            if len(row['행정구역']) <= 2: 
                dispname += row['행정구역'][-1]
        else:
            dispname = row['행정구역'][:-1] # 행정구역에 마지막 글자 제거
            
        #서대문구, 서귀포시 같이 이름이 3자 이상이면 작은 글자로 표시
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2
            
            plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight = 'bold',fontsize = fontsize, 
                         ha = 'center', va = 'center', color = annocolor, linespacing = linespacing)

    #시도 경계를 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c = 'black', lw = 4)
        
    plt.gca().invert_yaxis()
    #plt.gca().set_aspect(1)
    plt.axis('off')
    
    cb = plt.colorbar(shrink = 1, aspect = 10)
    cb.set_label(datalabel)
    
    plt.tight_layout()
    plt.savefig('.\\data8\\' + 'blockMap_' + targetData + '.png')
    plt.show()
```

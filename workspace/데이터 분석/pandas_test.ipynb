```python
dust = pd.read_excel("./data/dust1.xlsx")
dust.head()
```

```python
dust1 = dust.drop(["지역", "망", "측정소코드"], axis=1)
dust1.head()
```

```python
# dust1.columns = ["..."]는 칼럼명을 전부 갱신할 때 쓰는 방법
# 아래는 일부만 바꾸고 싶을때
dust1.rename(columns={"측정소명": "name", "측정일시": "date", "주소": "address"}, inplace=True)
dust1.head()
```

```python
# 정수형 날짜형으로 고치기 (일단 정수를 스트링으로 고친 후에 날짜형으로 고쳐야 함)
dust1["date"] = dust1["date"].astype(str)
dust1["date"] = dust1["date"].str.slice(0, 8)
dust1["date"] = pd.to_datetime(dust1["date"], format="%Y%m%d")
```

```python
dust1.head()
```

```python
dust1["year"] = dust1["date"].dt.year
dust1["month"] = dust1["date"].dt.month
dust1["day"] = dust1["date"].dt.day
dust1.head()
```

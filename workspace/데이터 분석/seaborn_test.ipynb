```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn
```

```python
plt.rcParams["font.family"] = ["Malgun Gothic"]
plt.rcParams["axes.unicode_minus"] = False
```

```python
seaborn.get_dataset_names() # 사용할 수 있는 데이터 셋 목록
```

```python
titanics = seaborn.load_dataset("titanic")
titanics.head()
```

```python
tips = seaborn.load_dataset("tips")
tips.head()
```

```python
seaborn.relplot(x="total_bill", y="tip", data=tips)
```

```python
health = pd.read_excel("./data/health_screenings_2020_1000ea.xlsx")
health.head()
```

```python
seaborn.relplot(data=health, x="systolic", y="diastolic")
```

```python
seaborn.relplot(x="total_bill", y="tip", data=tips, hue="day")
```

```python
seaborn.countplot(data=tips, x="day", hue="time")
```

```python
current_palette = seaborn.color_palette()
seaborn.palplot(current_palette)
plt.show()
```

```python
seaborn.palplot(seaborn.color_palette("pastel"))
plt.show()
```

```python
seaborn.set_palette("pastel")
seaborn.countplot(data=tips, x="day", hue="time")
```

```python
colors = ["Blue", "Red", "Green", "Orange"]
seaborn.palplot(seaborn.color_palette(colors))
plt.show()
```

```python
seaborn.set_palette(colors)
seaborn.countplot(data=tips, x="day", hue="time")
```

```python
seaborn.set_palette("muted")
seaborn.relplot(x="total_bill", y="tip", data=tips, hue="size", size="size")
```

```python
seaborn.relplot(data=tips, x="total_bill", y="tip", col="smoker", row="sex")
```

```python
seaborn.relplot(data=tips, x="total_bill", y="tip", col="day", col_wrap=2)
```

```python
# x = 수축 혈압, y = 이완 혈압, hue = 성별, col = drinking, row = smoking
seaborn.relplot(data=health, x="systolic", y="diastolic", hue="gender", col="smoking", row="drinking")
```

```python
data6 = health.loc[:, ['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]
data6.head()
```

```python
data6.loc[data6['gender']==1, ['gender']] = "Male"
data6.loc[data6['gender']==2, ['gender']] = "Female"
data6.loc[data6['drinking']==0, ['drinking']] = "Non-drinking"
data6.loc[data6['drinking']==1, ['drinking']] = "Drinking"
data6.loc[(data6['smoking']==1)|(data6['smoking']==2), ['smoking']] = "Non-smoking"
data6.loc[data6['smoking']==3, ['smoking']] = "Smoking"
data6.head()
```

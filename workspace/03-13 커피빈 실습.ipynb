```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
def CoffeeBean(lst):
    driver = webdriver.Chrome()
    driver.get("https://www.coffeebeankorea.com/store/store.asp")
    driver.implicitly_wait(10)

    storeList = driver.find_elements(By.CSS_SELECTOR, "#storeListUL > li")

    for store in storeList:
        distance = store.find_element(By.CSS_SELECTOR, "p.name > span > strong")
        name = store.find_element(By.CSS_SELECTOR, "p.name > span").text.replace(distance.text, "")
        address = store.find_element(By.CSS_SELECTOR, "p.address > span").text
        tel = store.find_element(By.CSS_SELECTOR, "p.tel > a").get_attribute("href").replace("tel:", "")
        lst.append((name, address, tel))
    
    driver.quit()
```

```python
result = []
CoffeeBean(result)

df = pd.DataFrame(result, columns=["점포명", "주소", "전화번호"])
df
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import urllib.request as req
import os
```

```python
driver = webdriver.Chrome()
driver.get("https://www.google.com/imghp")
driver.implicitly_wait(5)

query = "카피바라"
search = driver.find_element("name", "q")
search.send_keys(query)
search.submit()
driver.implicitly_wait(5)

# 스크롤 이동
new_height = 0
last_height = driver.execute_script("return window.scrollY")    # 현재 페이지의 세로 스크롤의 위치
PAUSE = 2
prev_height = driver.execute_script("return document.body.scrollHeight")    # 현재 페이지 전체 스크롤 높이

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(2)
    curr_height = driver.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height: break
    prev_height = curr_height

save_path = f"C:/JHM/BIGDATA/data/{query}"
if not os.path.exists(save_path):
    os.makedirs(save_path)  # os path에 save_path 경로 디렉토리가 없으면 하나 만들어라는 명령임

imgs = []
img_elements = driver.find_elements(By.CSS_SELECTOR, "div.H8Rx8c img.YQ4gaf")
for index, img_element in enumerate(img_elements):
    try:
        img_src = img_element.get_attribute("src")
        img_alt = img_element.get_attribute("alt")
        
        if img_src and "http" in img_src:
            imgs.append({"src": img_src, "alt": img_alt})
            req.urlretrieve(img_src, f"{save_path}/{query}_{index}.png")
            print(f"{save_path}/{query}_{index}.png", "다운로드됨...")
    except: pass

print("다운로드 완료.")

driver.close()
```

```python
# 실습

import re

url = "https://www.yes24.com/Product/Search?domain=ALL&query=%25ED%258C%258C%25EC%259D%25B4%25EC%258D%25AC"

html_doc = requests.get(url).text
soup = bs(html_doc, "html.parser")

books = []

items = soup.find_all(class_=re.compile("info$"))
for item in items:
    if item.find(class_="gd_name") is not None:
        name = item.find(class_="gd_name").text
    else:
        name = None
    
    if item.select_one("div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a") is not None:
        pub_group = item.select_one("div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a").text
    else:
        pub_group = None

    if item.select_one(".info_row.info_price > .yes_b") is not None:
        price = item.select_one(".info_row.info_price > .yes_b").text
    else:
        price = None

    if item.select_one(".rating_grade > .yes_b") is not None:
        rating = item.select_one(".rating_grade > .yes_b").text
    else:
        rating = None

    books.append([name, pub_group, price, rating])

for book in books:

    print(book)
```

```python
# 실습
import pandas as pd

url = "https://www.yes24.com/Product/Search?domain=ALL&query=%25ED%258C%258C%25EC%259D%25B4%25EC%258D%25AC"

webpage = requests.get(url).text
soup = bs(webpage, "html.parser")

books = []

items = soup.select("#yesSchList > li")
for item in items:
    if item.find(class_="gd_name") is not None:
        title = item.find(class_="gd_name").text
    else:
        title = None

    if item.select_one("div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a") is not None:
        author = item.select_one("div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a").text
    else:
        author = None

    if item.select_one("div.item_info > div.info_row.info_pubGrp > span.authPub.info_pub > a") is not None:
        publish = item.select_one("div.item_info > div.info_row.info_pubGrp > span.authPub.info_pub > a").text
    else:
        publish = None

    if item.select_one("div.item_info > div.info_row.info_price > strong > em") is not None:
        price = int(item.select_one("div.item_info > div.info_row.info_price > strong > em").text.replace(",", ""))
    else:
        price = 0

    if item.select_one("div.item_info > div.info_row.info_rating > span.rating_grade > em") is not None:
        grade = float(item.select_one("div.item_info > div.info_row.info_rating > span.rating_grade > em").text)
    else:
        grade = 0.0
    
    books.append((title, author, publish, price, grade))

df = pd.DataFrame(books, columns=["title", "author", "publish", "price", "grade"])
df.to_csv("../data/books.csv", index=False)

df
```
